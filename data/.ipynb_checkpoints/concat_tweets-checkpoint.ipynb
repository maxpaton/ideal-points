```python
import numpy as np
import pandas as pd
import os
import nltk
from nltk.stem import WordNetLemmatizer
```

## Load

```python
out_path = '../tbip/data/covid-tweets-2020/raw/covid_tweets.csv'
base_path = 'tweets/'

ls = []
for path, directory, file in os.walk(base_path):
    for name in sorted(file):
        if name.endswith('.csv'):
            filename = os.path.join(path, name)
            df = pd.read_csv(filename, header=0, index_col=None, engine='python')
            ls.append(df)

tweets = pd.concat(ls, axis=0, ignore_index=True)
print(len(tweets))
# tweets.to_csv(out_path)
```

```python
tweets.head()
```

## Preprocess

```python
# drop NaNs
tweets = tweets.dropna(subset=['description', 'tweet'])
# remove Retweeted content
tweets = tweets.drop(tweets[tweets.tweet.str.startswith('RT')].index)
print(len(tweets))

def preprocess(tweet):
    parsed = nltk.word_tokenize(tweet)
```

```python
lemmatizer = WordNetLemmatizer()

academic_kw = set(['prof', 'professor' 'lecture', 'lecturer' 'university', 'school', 'department', 'edu', 'researcher', 'doctorate',
              'faculty', 'academic', 'college', 'post', 'doctoral', 'phd', 'ph.d', 'dphil', 'teacher',
              'conference', 'staff', 'teaching', 'academy', 'taught', 'academical', 'scholar'])
academic_kw = set([lemmatizer.lemmatize(w) for w in academic_kw])

journalist_kw = set(['digital', 'media', 'multimedia', 'journal', 'journalist', 'news', 'journalism', 'digital',
                 'reporter', 'writer', 'reporting', 'editor', 'newspaper', 'blogger', 'columnist', 
                'correspondent', 'press', 'writing'])
journalist_kw = set([lemmatizer.lemmatize(w) for w in journalist_kw])

doctors_ls = 'physician vet surgeon medicine medico doc doctor md veterinarian gastroenterologist allergist dr. medical patient neurologist neuro neuroscientist neuroscience pediatrician medication clinic hospital gp specialist veterinary practitioner practitioner surgeon sophisticate dermatologist oncologist obstetrician medic clinician cardiologist internist nephrologist gynecologist radiologist nurse care dentist therapist pharmacist psychologist doctoral pathologist dr healer doctorate theologian radiotherapy neuropsychologist medician biomedicine geriatrician virologist disease virus oncology psychotic clinical placebo gastroenterology dermatology pseudomedical biomedical biomed'
doctors_kw = set(doctors_ls.split(' '))
doctors_kw = set([lemmatizer.lemmatize(w) for w in doctors_kw])

politicians_ls = 'minister mp leader party parliament house representative decocrat commons lords politics leader governor democrat candidate legislator demagogue campaigner statesman politico senator socialist republican political leader political presidency technocrat labourite democratic lawmaker jurist parliamentarian journalist servant reformer nationalist congressman populist activist diplomat stateswoman federalist communist liberal conservative labour wing member politically president tory government served presidential advocate industrialist minister governmental parliamentary deputy'
politicians_kw = set(politicians_ls.split(' '))
politicians_kw = set([lemmatizer.lemmatize(w) for w in politicians_kw])
```

```python
tweets['description'] = tweets.description.apply(lambda x: nltk.word_tokenize(x))
tweets['description'] = tweets.description.apply(lambda x: [w.lower() for w in x])
tweets['description'] = tweets.description.apply(lambda x: [lemmatizer.lemmatize(w) for w in x])
```

```python
tweets.head()
```

```python
tweets['academic_count'] = tweets.description.apply(lambda x: len([w for w in x if w in academic_kw]))
tweets['journalist_count'] = tweets.description.apply(lambda x: len([w for w in x if w in journalist_kw]))
tweets['doctor_count'] = tweets.description.apply(lambda x: len([w for w in x if w in doctors_kw]))
tweets['politician_count'] = tweets.description.apply(lambda x: len([w for w in x if w in politicians_kw]))
```

```python
tweets.head(10)
```

```python
counts = np.array([tweets.academic_count, tweets.journalist_count, tweets.doctor_count, tweets.politician_count]).T
sums = tweets.academic_count + tweets.journalist_count + tweets.doctor_count + tweets.politician_count
proba = counts/sums.values.reshape(-1,1)
tweets['proba'] = proba.tolist()
```

```python
tweets
```

```python
# choose label based on max author type count
mask = (pd.notnull(tweets.proba.apply(lambda x: x[0]))) & (tweets.proba.apply(lambda x: x.count(max(x))==1))
valid = tweets[mask]
tweets['label'] = 0
tweets.loc[mask, 'label'] = tweets[['academic_count', 'journalist_count', 'doctor_count', 'politician_count']].idxmax(1)

# encode categorical author type label
# d = {'academic_count': 1, 'journalist_count': 2, 'doctor_count': 3, 'politician_count': 4}
d = {'academic_count': 'academic', 'journalist_count': 'journalist', 'doctor_count': 'doctor', 'politician_count': 'politician'}
tweets['label'] = tweets.label.apply(lambda x: d[x] if x != 0 else x)
```

## Write dataframe to csv in correct format

```python
tweets = tweets.loc[tweets.label != 0]
tweets = tweets[['label', 'id', 'time', 'tweet']]
tweets.columns = ['screen_name', 'id', 'created_at', 'text']
print(len(tweets))
tweets.to_csv('../tbip/data/covid-tweets-2020/raw/tweets.csv', index=False)
```
